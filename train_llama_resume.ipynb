```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset
```

```python
model_path = "./llama" 
dataset = load_dataset('json', data_files='resume_data.jsonl')
#dataset
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_path)
```

```python
def tokenize_function(example):
    print("Debug:", example)  # Debugging line to check the data type of prompt and response
    combined = example["prompt"] + " " + example["response"]
    return tokenizer(combined, padding="max_length", truncation=True, max_length=256)
tokenized_dataset = dataset.map(tokenize_function, batched=True)
```
